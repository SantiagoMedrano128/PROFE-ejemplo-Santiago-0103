```python
# Clase automovil, atributos, marca, placas, precio
# funciones leerdatos(), mostrardatos()
# Crear tres objetos carro1, carro2, carro3

class Automovil:
    def __init__(self, marca, placas, precio):
        self.marca = marca
        self.placas = placas
        self.precio = precio

    def leerdatos(self):
        self.marca = input("Ingrese la marca del automovil: ")
        self.placas = input("Ingrese las placas del automovil: ")
        self.precio = float(input("Ingrese el precio del automovil: "))

    def mostrardatos(self):
        print("Marca:", self.marca)
        print("Placas:", self.placas)
        print("Precio:", self.precio)

print("Carro Numero 1")
carro1 = Automovil("", "", "") # Creando el objeto carro 1
carro1.leerdatos()
carro1.mostrardatos()
print("-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-")
print("Carro Numero 2")
carro2 = Automovil("", "", "") # Creando el objeto carro 2
carro2.leerdatos()
carro2.mostrardatos()
print("-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-")
print("Carro Numero 3")
carro3 = Automovil("", "", "") # Creando el objeto carro 3
carro3.leerdatos()
carro3.mostrardatos()
print("Programa realizado por = Santiago Medrano NL = 37")
```
